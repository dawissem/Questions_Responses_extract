```python
data= pd.read_excel('/content/drive/MyDrive/Patie IA PROJET PFE/TW-Tests Evaluation Métier TN - version original.xlsx', sheet_name="Quiz-Metier")
```

```python
data
```

```python
# Réinitialiser la liste des questions et réponses
quiz_data = []
current_question = None
question_type = None

# Parcourir les lignes du DataFrame
for index, row in data.iterrows():
    col_1 = row.iloc[1]  # Colonne des numéros de question (peut contenir NaN)
    question_col = str(row.iloc[2]) if pd.notna(row.iloc[2]) else ""  # Colonne des questions
    answer_col = str(row.iloc[3]) if pd.notna(row.iloc[3]) else ""  # Colonne des réponses

    # Vérifier si la ligne indique un type de question
    if "Question à multiple réponses" in question_col:
        question_type = " Choix multiple"
    elif "Réponse unique" in question_col:
        question_type = " Choix unique"

    # Vérifier si la ligne contient une question (col_1 a une valeur numérique)
    elif pd.notna(col_1) and isinstance(col_1, (int, float)):
        if current_question:  # Ajouter la question précédente avant d'en commencer une nouvelle
            quiz_data.append(current_question)
        current_question = {
            "number": f"**{len(quiz_data) + 1}**",  # Numéro en gras
            "type": question_type.capitalize() if question_type else "Reponse user",  # Type de question
            "question": question_col.strip(),
            "answers": {}
        }

    # Vérifier si la ligne contient une réponse (col_1 vide et answer_col non vide)
    elif current_question and answer_col and pd.isna(col_1):
        answer_index = len(current_question["answers"]) + 1  # Numérotation des réponses
        current_question["answers"][str(answer_index)] = answer_col.strip()

# Ajouter la dernière question trouvée
if current_question:
    quiz_data.append(current_question)

# Convertir en JSON et afficher
quiz_json = json.dumps(quiz_data, ensure_ascii=False, indent=4)
quiz_json
```

```python
data_responses = pd.read_excel('/content/drive/MyDrive/Patie IA PROJET PFE/TW-Tests Evaluation Métier TN - version original.xlsx', sheet_name="Rep-Quiz-Metier")
```

```python
data_responses
```

```python
# Réinitialiser la liste des réponses par question
responses_data = []
current_question = None

# Parcourir les lignes du DataFrame
for index, row in data_responses.iterrows():
    col_1 = row.iloc[1]  # Colonne des numéros de question (peut contenir NaN)
    question_col = str(row.iloc[2]) if pd.notna(row.iloc[2]) else ""  # Colonne des questions
    answer_col = str(row.iloc[3]) if pd.notna(row.iloc[3]) else ""  # Colonne des réponses
    weight_col = row.iloc[4] if pd.notna(row.iloc[4]) else 0  # Pondération
    score_col = row.iloc[5] if pd.notna(row.iloc[5]) else 0  # Score de la réponse
    total_score_col = row.iloc[6] if pd.notna(row.iloc[6]) else None  # Score total de la question

    # Vérifier si la ligne contient une nouvelle question (numéro de question présent)
    if pd.notna(col_1) and isinstance(col_1, (int, float)):
        if current_question:  # Ajouter la question précédente avant d'en commencer une nouvelle
            responses_data.append(current_question)
        current_question = {
            "number": f"{int(col_1)}",  # Numéro en gras
            "question": question_col.strip(),
            "total_score": None,  # À mettre à jour plus tard
            "answers": []
        }

    # Vérifier si la ligne contient une réponse associée à la question en cours
    elif current_question and answer_col:
        current_question["answers"].append({
            "answer": answer_col.strip(),
            "pondération": weight_col,
            "score": score_col
        })

    # Vérifier si la ligne contient le score total de la question
    elif current_question and total_score_col is not None:
        current_question["total_score"] = total_score_col

# Ajouter la dernière question trouvée
if current_question:
    responses_data.append(current_question)

# Convertir en JSON et afficher
responses_json = json.dumps(responses_data, ensure_ascii=False, indent=4)
responses_json
```
